```python
import pandas as pd
```

```python
review = pd.read_csv('amazon-video-game-review-clean.csv')
```

```python
review.head()
```

```python
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
```

```python
review['reviewText'] = review['reviewText'].astype(str).str.lower()
review.head(3)
```

```python
# tokenize
regexp = RegexpTokenizer('\w+')
review['token']=review['reviewText'].apply(regexp.tokenize)
review.head(3)
```

```python
# remove stopwords
stopwords = nltk.corpus.stopwords.words("english")
review['token'] = review['token'].apply(lambda x: [item for item in x if item not in stopwords])
review.head(3)
```

```python
# remove infrequent words
review['string'] = review['token'].apply(lambda x: ' '.join([item for item in x if len(item) > 2]))
all_words = ' '.join([word for word in review['string']])
tokenized_words = nltk.tokenize.word_tokenize(all_words)
```

```python
# nltk.download('punkt')
```

```python
from nltk.probability import FreqDist

fdist = FreqDist(tokenized_words)
fdist
```

```python
review['frequent'] = review['token'].apply(lambda x: ' '.join([item for item in x if fdist[item] >= 1]))
```

```python
review.head()
```

```python
# Lemmatization
# nltk.download('wordnet')
# nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
wordnet_lem = WordNetLemmatizer()

review['lemmatization'] = review['frequent'].apply(wordnet_lem.lemmatize)
```

```python
sum(review['lemmatization']!=review['frequent'])
```

```python
# Sentiment
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
review['polarity'] = review['lemmatization'].apply(lambda x: analyzer.polarity_scores(x))
review.head(3)
```

```python
review_sentiment = pd.concat(
    [review.drop(['polarity'], axis=1), 
     review['polarity'].apply(pd.Series)], axis=1)
# review_sentiment['nltk_sentiment'] = review_sentiment['compound'].apply(lambda x: 'positive' if x >0 else 'neutral' if x==0 else 'negative')
```

```python
review_sentiment = review_sentiment.drop(['token', 'string', 'frequent', 'neg', 'neu', 'pos'], axis=1)
```

```python
review_sentiment.rename(columns={'compound': 'nltk', 'lemmatization': 'cleanedText'}, inplace=True)
```

```python
review_sentiment.to_csv('review-sentiment.csv', index=False)
```
